```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/a10.csv', parse_dates=['date'])
```

```python
df.head()
```

```python
df['year'] = [i.year for i in df.date]
```

```python
df.head()
```

```python
df['month'] = [i.strftime('%b') for i in df.date]
df.head()
```

```python
years = df['year'].unique()
```

```python
import random
hexadecimal_alphabets = '0123456789ABCDEF'
my_colors = ["#" + ''.join([random.choice(hexadecimal_alphabets) for j in
range(6)]) for i in range(24)]
```

```python
plt.figure(figsize=(18,18))
for i in range(len(years)):
  plt.plot('month','value',data=df.loc[df.year==years[i],:],color=my_colors[i],linewidth=5,label=years[i])
plt.legend()
```
