```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/a10.csv', parse_dates=['date'], index_col='date')
```

```python
plt.plot(df)
```

```python
res_add=seasonal_decompose(df['value'],model='additive',extrapolate_trend='freq')
res_mul=seasonal_decompose(df['value'],model='multiplicative',extrapolate_trend='freq')
```

```python
res_add.plot()
```

```python
res_mul.plot()
```

```python
res_add.resid.plot(color='tab:red')
res_mul.resid.plot(color='blue')
```
